```python
list1 =[1,3,4,5,7]
list2=[1,2,3,4,5]

# Convert lists to sets to use the difference method
set1 = set(list1)
set2 = set(list2)

# Find the difference between the sets
difference = set1.difference(set2)

# Convert the resulting set back to a list if needed
difference_list = list(difference)
print(difference_list)  # Output: [7]
```

### **Dict Comprehension**

```python
dictionary={"a":1,
            "b":2,
            "c":3,
            "d":4}

{k:v **2 for (k,v) in dictionary.items()}
```
