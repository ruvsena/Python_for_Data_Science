```python
#kategorik:sütun grafiği ve countplot
#sayısal değişken: hist boxplot


import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
pd.set_option("display.max_columns",None)
pd.set_option("display.max_rows",200)
df=sns.load_dataset("titanic")

df.head()
```

```python
#df["sex"].value_counts()

df["sex"].value_counts().plot(kind='bar')
plt.show()
```

```python
plt.hist(df["age"])
plt.show()
```

```python
plt.boxplot(df["fare"])
plt.show()
```

```python
import numpy as np
x = np.array([1,8])
y = np.array([0,150])
plt.plot(x,y)


plt.plot(x,y,'o')
plt.show()
```

```python
# line

y= np.array([13,28,11,101])
plt.plot(y,linestyle='dotted',color='r')#dashdot,
plt.show()
```

```python
x=np.array([23,18,31,10])
y=np.array([13,28,11,101])
plt.plot(x)
plt.plot(y)
plt.show
```

```python
#labels

x=np.array([23,18,31,10])
y=np.array([13,28,11,101])
plt.plot(x)
plt.plot(y)

plt.title("ana başlık grafik")
plt.xlabel("x ekseni")
plt.ylabel("y ekseni")
plt.grid()
plt.show

```

## **Seaborn ile veri görselleştirme**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df=sns.load_dataset("tips")

df.head()
```

```python
df["sex"].value_counts()
sns.countplot(x=df["sex"],data=df)
plt.show()
```

```python
sns.boxplot(x=df["total_bill"])
plt.show()
```
